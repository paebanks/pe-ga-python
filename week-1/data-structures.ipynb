```python
#lists
#zero indexed

names = [] #creates an empty list

days_of_week = ['Mon',  'Tue', 'Wed', 'Thu', 'Fri', 'Sat']

colors = list() #using the list constructor, also creates an empty list

#adding to a list
names.append('Pavel') #append adds to end of list

print(names)

#add your name to the list
names.append('Enzo')
names.append('Sanjeev')
names.append('Ella')
names.append('Son')
names.append('Priscilla')

#print the list
print(names)

#how many names are in the list?
no_of_names = len(names)
print(no_of_names)


#remove an item from end of list
last_in_list = names.pop()

print(names)
print(last_in_list)

#remove from a specific index
second_in_list = names.pop(1)
print(second_in_list)
print(names)

#insert at any position
names.insert(1, second_in_list)
print(names)

#count the occurence of an item in a list

numbers = [1, 3, 5, 3, 4, 5, 3, -1, 5]

number_of_fives = numbers.count(5)
print(number_of_fives)

#sum my numbers
total = sum(numbers)
print(total)
```
